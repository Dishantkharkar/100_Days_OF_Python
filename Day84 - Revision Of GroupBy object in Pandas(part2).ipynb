# **Revision of GroupBy Attributes and Methods in Pandas**

* len - find total number of groups
* size - find items in each group
* nth item - first()/nth/last items
* get_group - vs filtering
* groups
* describe
* sample
* nunique
* agg method
* apply -> builtin function

```python
import numpy as np
import pandas as pd

movies = pd.read_csv('Data\Day35\imdb-top-1000.csv')
```

```python
genres = movies.groupby('Genre')
```

### 1. len

```python
len(movies.groupby('Genre'))
```

### 2. nunique

```python
movies['Genre'].nunique()
```

### 3. size

```python
movies.groupby('Genre').size()
```

### 4. nth

```python
genres = movies.groupby('Genre')
# genres.first()
# genres.last()
genres.nth(6)
```

### 5. get_group

```python
genres.get_group('Fantasy')
```

```python
# with simple but its slow
movies[movies['Genre'] == 'Fantasy']
```

### 6. describe 

```python
genres.describe()
```

### 7. sample

```python
genres.sample(2,replace=True)
```

### 8. nunique()

```python
genres.nunique()
```

### 9. agg method

```python
# passing dict
genres.agg(
    {
        'Runtime':'mean',
        'IMDB_Rating':'mean',
        'No_of_Votes':'sum',
        'Gross':'sum',
        'Metascore':'min'
    }
)
```

```python
genres.agg(
    {
        'Runtime':['min','mean'],
        'IMDB_Rating':'mean',
        'No_of_Votes':['sum','max'],
        'Gross':'sum',
        'Metascore':'min'
    }
)
```

### 10. apply -> builtin function

```python
genres.apply(min)
```
