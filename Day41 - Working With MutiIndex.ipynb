# **Working with MultiIndex DataFrames**

```python
import numpy as np
import pandas as pd
```

### **Creating MutiIndex Dataframe**


```python
index_val = [('cse',2019),('cse',2020),('cse',2021),('cse',2022),('ece',2019),('ece',2020),('ece',2021),('ece',2022)]
multiindex = pd.MultiIndex.from_tuples(index_val)
multiindex.levels
```

```python
branch_df = pd.DataFrame(
    [
        [1,2,0,0],
        [3,4,0,0],
        [5,6,0,0],
        [7,8,0,0],
        [9,10,0,0],
        [11,12,0,0],
        [13,14,0,0],
        [15,16,0,0],
    ],
    index = multiindex,
    columns = pd.MultiIndex.from_product([['delhi','mumbai'],['avg_package','students']])
)

branch_df
```

#### How to Extract

```python
# extracting single row
branch_df.loc[('cse',2022)]
```

```python
branch_df
```

```python
# extract multiple rows
branch_df.loc[('cse',2021):('ece',2021)]
```

```python
# using iloc
branch_df.iloc[2:5]
```

```python
branch_df.iloc[2:8:2]
```

```python
# extacting cols
branch_df['delhi']['students']
```

```python
branch_df.iloc[:,1:3]
```

```python
# Extracting both
branch_df.iloc[[0,4],[1,2]]
```

### Sorting


```python
branch_df.sort_index(ascending=False)
```

```python
branch_df.sort_index(ascending=[False,True])
```

```python
branch_df.sort_index(level=0,ascending=[False])
```

```python
# multiindex dataframe(col) -> transpose
branch_df.transpose()
```

```python
# swaplevel
branch_df.swaplevel(axis=1)
```

```python
branch_df.swaplevel()
```
